## 2 Gaussian Classiﬁcation

Let $f (x | C_i) \sim N(\mu_i, \mu_2)$ for a two-class, one-dimensional classification problem with classes $C_1$ and $C_2$, $P(C_1) = P(C_2) = 1/2$, and $\mu_2 > \mu_1$.

1. Find the Bayes optimal decision boundary and the corresponding Bayes decision rule.


$$
P(Y=C_i|X=x)=\frac{P(X=x|Y=C_i)P(Y=C_i)}{P(X=x)}
$$

$$
f_{C_i}(x)=\frac{1}{\sqrt{2\pi}\sigma}exp\{-\frac{||x-\mu_i||^2}{2\sigma^2}\}
$$

$$
Q_{C_i}(x)=\ln(\sqrt{2\pi}f_{C_i}(x)\pi_{C_i})=-\frac{||x-\mu_i||^2}{2\sigma^2}-\ln\sigma-\ln2
$$

$$
Q_{C_1}(x)>Q_{C_2}(x)
$$



Bayes decision rule:
$$
\begin{equation}
r^*(x)=
\left\{
             \begin{array}{lr}
            C_1, & x<\frac{\mu_1+\mu_2}{2} \\
            C_2 & o.w.
             \end{array}
\right.
\end{equation}
$$


2.	The Bayes error (the probability of misclassification)

$$
P_e=\frac{1}{2}\int_{-\infty}^{\frac{\mu_1+\mu_2}{2}}f_{C_2}(x)+\frac{1}{2}\int_{\frac{\mu_1+\mu_2}{2}}^{\infty}f_{C_1}(x)=\frac{1}{\sqrt{2\pi}}\int_{\frac{\mu_2-\mu_1}{2\sigma}}^{\infty}e^{-z^2/2}dz
$$

## 3 Isocontours of Normal Distributions

Let $f (\mu, \Sigma)$ be the probability density function of a normally distributed random variable in $\mathbf{R^2}$. Write code to plot the isocontours of the following functions, each on its own separate figure. You’re free to use any plotting libraries available in your programming language; for instance, in Python you can use Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math

def gaussian_pdf(mean_matrix, cov_matrix, x, y):
    pi = np.pi
    pos = np.array([x, y])
    cov_matrix_det = np.linalg.det(cov_matrix)
    cov_matrix_inv = np.linalg.inv(cov_matrix)
    g = 1/(2*pi*math.sqrt(cov_matrix_det)) * math.exp(-0.5*(pos-mean_matrix).T.dot(cov_matrix_inv.dot(pos-mean_matrix)))
    return g
    
```

1. $f (\mu, \Sigma)$, where $\mu= \begin{bmatrix} 1  \\ 1  \end{bmatrix}$ and $\Sigma=\begin{bmatrix} 1 & 0 \\ 0 & 2    \end{bmatrix}$.

```python
xi, yi = np.linspace(-2, 4, 100), np.linspace(-2, 4, 100)
xi, yi = np.meshgrid(xi, yi)
mean_matrix = np.array([1, 1])
cov_matrix = np.array([[1, 0], [0, 2]])
z = np.array([gaussian_pdf(mean_matrix, cov_matrix, x, y) for (x,y) in zip(xi.ravel(), yi.ravel())]).reshape(xi.shape)


fig = plt.figure(figsize=(10, 5))

plot1 = fig.add_subplot(1, 2, 1)
plot1.set_aspect('equal')
plot1.contourf(xi,yi,z,10,cmap='Oranges')
plot1_contour = plot1.contour(xi,yi,z,10, colors='black',linewidths=1)
plt.clabel(plot1_contour, inline = True, fontsize = 10)


plot2 = fig.add_subplot(1, 2, 2, projection='3d')
plot2.plot_surface(xi,yi,z,cmap='Oranges')

plt.show()
```

2. $f (\mu, \Sigma)$, where $\mu= \begin{bmatrix} -1  \\ 2  \end{bmatrix}$ and $\Sigma=\begin{bmatrix} 2 & 1 \\ 1 & 4    \end{bmatrix}$.

```python
xi, yi = np.linspace(-4, 2, 100), np.linspace(-2, 4, 100)
xi, yi = np.meshgrid(xi, yi)
mean_matrix = np.array([-1, 2])
cov_matrix = np.array([[2, 1], [1, 4]])
z = np.array([gaussian_pdf(mean_matrix, cov_matrix, x, y) for (x,y) in zip(xi.ravel(), yi.ravel())]).reshape(xi.shape)


fig = plt.figure(figsize=(10, 5))

plot1 = fig.add_subplot(1, 2, 1)
plot1.set_aspect('equal')
plot1.contourf(xi,yi,z,10,cmap='Oranges')
plot1_contour = plot1.contour(xi,yi,z,10, colors='black',linewidths=1)
plt.clabel(plot1_contour, inline = True, fontsize = 10)


plot2 = fig.add_subplot(1, 2, 2, projection='3d')
plot2.plot_surface(xi,yi,z,cmap='Oranges')

plt.show()
```

## 4 Eigenvectors of the Gaussian Covariance Matrix

Consider two one-dimensional random variables $X_1 \sim \mathcal{N}(3, 9)$ and $X_2 ∼ \frac{1}{2} X_1 + \mathcal{N}(4, 4)$, where $\mathcal{N}(\mu, \sigma^2)$ is a Gaussian distribution with mean $\mu$ and variance $\sigma^2$. Write a program that draws $n = 100$ random two-dimensional sample points from $(X_1, X_2)$ such that the $i$th value sampled from $X_2$ is calculated based on the ith value sampled from $X_1$. In your code, make sure to choose and set a fixed random number seed for whatever random number generator you use, so your simulation is reproducible, and document your choice of random number seed and random number generator in your write-up. For each of the following parts, include the corresponding output of your program.

```python
mean_1 = 3
mean_2 = 4
std_1 = 3
std_2 = 2
n = 100
np.random.seed(0)
X_1 = np.random.normal(mean_1, std_1, n)
X_2 = np.random.normal(mean_2, std_2, n)
X_2 += X_1/2
```

(a)	Compute the mean (in $\mathbf{R^2}$) of the sample.

```python
print(X_1.mean())
print(X_2.mean())
```

(b)	Compute the 2 × 2 covariance matrix of the sample.

```python
X = np.vstack((X_1, X_2))
cov_matrix = np.cov(X)
print(cov_matrix)
```

(c)	Compute the eigenvectors and eigenvalues of this covariance matrix.

```python
w, v = np.linalg.eig(cov_matrix)
print(w)
print(v)
```

(d)	On a two-dimensional grid with a horizonal axis for $X_1$ with range [−15, 15] and a vertical axis for $X_2$ with range [−15, 15], plot all $n = 100$ data points, and arrows representing both covariance eigenvectors. The eigenvector arrows should orig-inate at the mean and have magnitudes equal to their corresponding eigenvalues.


```python
fig, ax = plt.subplots()
plt.grid()
ax.scatter(X_1, X_2)
ax.axis(xmin=-15,xmax=15)
ax.axis(ymin=-15,ymax=15)
ax.set_aspect('equal')

v1 = v[:, 0].dot(w[0])
v2 = v[:, 1].dot(w[1])
ax.arrow(X_1.mean(), X_2.mean(), v1[0], v1[1], head_width=1, head_length=1, fc='g', ec='g')
ax.arrow(X_1.mean(), X_2.mean(), v2[0], v2[1], head_width=1, head_length=1, fc='r', ec='r')
```

(e) Let $U = [v_1,v_2]$ be a 2 × 2 matrix whose columns are the eigenvectors of the covariance matrix, where $v_1$ is the eigenvector with the larger eigenvalue. We use $U^T$ as a rotation matrix to rotate each sample point from the $(X_1, X_2)$ coordinate system to a coordinate system aligned with the eigenvectors. (As $U^T=U^{-1}$, the matrix U reverses this rotation, moving back from the eigenvector coordinate system to the original coordinate system). Center your sample points by subtracting the mean µ from each point; then rotate each point by $U^T$, giving $x_{rotated} = U^T(x - \mu)$. Plot these rotated points on a new two dimensional-grid, again with both axes having range [−15, 15].

Supposed that:
$$
f(x)=\frac{1}{(\sqrt{2\pi})^d\sqrt{\Sigma_I}}exp\{-\frac{1}{2}x^T\Sigma_I^{-1}x\},\Sigma_I=I
$$
Apply the transform:
$$
t=Ax=RSx,x=A^{-1}t
$$

$$
t'=t+\mu
$$



in which, $S$ is a scale matrix, $R$ is a rotation matrix. We get:
$$
f(t')\sim exp\{-\frac{1}{2}(t'-\mu)^T(AA^T)^{-1}(t'-\mu)\}
$$
Obviously, the covariance matrix:
$$
\Sigma=AA^T
$$
Apply the eigendecomposition:
$$
\Sigma=U\Lambda U^T=(U\Lambda^{1/2})(U\Lambda^{1/2})^T
$$
We can get:
$$
A=U\Lambda^{1/2}=RS
$$

$$
R=U
$$

$$
S=\Lambda^{1/2}
$$

Now, let's apply the inverse transform:

(1) centering:
$$
t=t'-\mu
$$
(2) decorrelating:
$$
R^{-1}t=U^{-1}t=U^Tt
$$
(3) sphering:
$$
x=S^{-1}U^Tt=\Lambda^{-1/2}U^Tt
$$


```python
U = v
P = np.diag(1/np.sqrt(w))
```

```python
# centering
X_1 -= X_1.mean()
X_2 -= X_2.mean()
fig, ax = plt.subplots()
plt.grid()
ax.scatter(X_1, X_2)
ax.axis(xmin=-15,xmax=15)
ax.axis(ymin=-15,ymax=15)
ax.set_aspect('equal')
```

```python
# decorrelating
X = v.T.dot(np.array([X_1, X_2]))
X_1 = X[0, :]
X_2 = X[1, :]
fig, ax = plt.subplots()
plt.grid()
ax.scatter(X_1, X_2)
ax.axis(xmin=-15,xmax=15)
ax.axis(ymin=-15,ymax=15)
ax.set_aspect('equal')
```

```python
# sphering
X = P.dot(np.array([X_1, X_2]))
X_1 = X[0, :]
X_2 = X[1, :]
fig, ax = plt.subplots()
plt.grid()
ax.scatter(X_1, X_2)
ax.axis(xmin=-15,xmax=15)
ax.axis(ymin=-15,ymax=15)
ax.set_aspect('equal')
```
