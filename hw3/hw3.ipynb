## 3 Isocontours of Normal Distributions

Let $f (\mu, \Sigma)$ be the probability density function of a normally distributed random variable in $\mathbf{R^2}$. Write code to plot the isocontours of the following functions, each on its own separate figure. You’re free to use any plotting libraries available in your programming language; for instance, in Python you can use Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math

def gaussian_pdf(mean_matrix, cov_matrix, x, y):
    pi = np.pi
    pos = np.array([x, y])
    cov_matrix_det = np.linalg.det(cov_matrix)
    cov_matrix_inv = np.linalg.inv(cov_matrix)
    g = 1/(2*pi*math.sqrt(cov_matrix_det)) * math.exp(-0.5*(pos-mean_matrix).T.dot(cov_matrix_inv.dot(pos-mean_matrix)))
    return g
    
```

## 6 Maximum Likelihood Estimation
Let $X_1, \cdots, X_n \in R^d$ be n sample points drawn independently from a multivariate normal distribution $\mathcal{N}(\mu, \Sigma)$.

(a) Suppose the normal distribution has an unknown diagonal covariance matrix:

$$
\Sigma=
\begin{bmatrix}
    \sigma_1^2 &            &        &            \\
               & \sigma_2^2 &        &            \\
               &            & \ddots &            \\
               &            &        & \sigma_d^2 \\
\end{bmatrix}
$$
and an unknown mean $\mu$. Derive the maximum likelihood estimates, denoted $\hat{\mu}$ and $\hat{\sigma}_i$, for $\mu$ and $\sigma_i$. Show all your work.




$$
f(X_i)=\frac{1}{(\sqrt{2\pi})^d\sqrt{|\Sigma|}}\exp\{-\frac{1}{2}(X_i-\mu)^T\Sigma^{-1}(X_i-\mu)\}
$$
Likelihood:
$$
\mathcal{L}(\mu,\sigma;X_1,\cdots,X_n)=\prod_{i=1}^nf(X_i)
$$
Log likelihood:
$$
\mathcal{l}(\mu,\sigma;X_1,\cdots,X_n)=\sum_{i=1}^n\ln f(X_i)=\sum_{i=1}^n[-\frac{1}{2}(X_i-\mu)^T\Sigma^{-1}(X_i-\mu)-d\ln\sqrt{2\pi}-\ln\sqrt{|\Sigma|}]
$$
Take the derivative with respect to $\mu$ and set it to zero:
$$
\nabla_{\mu}\mathcal{l}=\sum_{i=1}^n\Sigma^{-1}(X_i-\mu)=0
$$

$$
\hat{\mu}=\frac{1}{n}\sum_{i=1}^nX_i
$$

To find $\hat{\sigma}_i$:
$$
f(X_i)=\frac{1}{(\sqrt{2\pi})^d\prod_{j=1}^d\sigma_j}\exp\{-\frac{1}{2}(X_i-\mu)^T\Sigma^{-1}(X_i-\mu)\}
$$

$$
\mathcal{l}(\mu,\sigma;X_1,\cdots,X_n)=\sum_{i=1}^n[-\frac{1}{2}\sum_{j=1}^d\frac{1}{\sigma_j^2}||X_i-\mu||^2-d\ln\sqrt{2\pi}-\ln\prod_{j=1}^d\sigma_j]
$$

$$
\frac{\partial\mathcal{l}}{\partial \sigma_i}=\sum_{i=1}^n\frac{||X_i-\mu||^2-\sigma_i^2}{\sigma_i^3}=0
$$

$$
\sigma_i^2=\frac{1}{n}\sum_{i=1}^n||X_i-\mu||^2=\frac{1}{n}\sum_{i=1}^n||X_i-\hat{\mu}||^2
$$

```python
class LDA:
    def __init__(self, labels):
        self.labels = labels
        self.means = {}
        self.cov = []
        self.P = {}
    
    def fit(self, X, y):
        n = len(y)
        cov = 0
        means = {}
        P = {}
        for label in self.labels:
            X_class = X[y.ravel()==label, :]
            n_c = X_class.shape[0]
            P[label] = n_c / n
            means[label] = X_class.mean(axis=0)
            cov += np.cov(X_class, rowvar=False) * n_c
        cov = cov / n
        self.means = means
        self.cov = cov
        self.P = P
        
    def linear_discriminant_fn(self, X, mean_c, P_c):
        cov_inv = np.linalg.pinv(self.cov)
        return mean_c.dot(cov_inv).dot(X.T) \
            - 1/2*mean_c.dot(cov_inv).dot(mean_c.T) \
            + np.log(P_c)
    
    def predict(self, X):
        max_probs = [0 for i in range(0, X.shape[0])]
        y = [0 for i in range(0, X.shape[0])]
        for label in self.means.keys():
            mean_c = means[label]
            P_c = self.P[label]
            probs = self.linear_discriminant_fn(X, mean_c, P_c)
            for i in range(0, len(probs)):
                if probs[i] > max_probs[i]:
                    max_probs[i] = probs[i]
                    y[i] = label
        return y
    
    def accuracy(self, X, y):
        y_p = self.predict(X)
        n = len(y)
        n_accuracy = (y_p == y.ravel()).sum()
        print(n_accuracy/n)
        return n_accuracy/n
```

```python
# https://github.com/chenyc15/Berkeley-CS289A/blob/master/hw3/hw3_code_part2.ipynb
class QDA:
    def __init__(self, a):
        self.Mean = 0
        self.Cov = 0
        self.P = 0
        self.a = a
        
    def Q(self, X, m, C, p):
        PreMat = np.linalg.inv(C)
        Const = - 1/2*np.log(np.linalg.det(C)+1e-20) + np.log(p)
        Qc = np.array([(-1/2*(x-m).dot(PreMat).dot(x-m)) + Const for x in X])
        return Qc
    
    def fit(self, X, y):
        N = len(y)
        d = X.shape[1]
        labels = set(y.ravel())
        Cov = np.zeros([X.shape[1], d, len(labels)])
        Mean = np.zeros([len(labels), d])
        P = np.zeros([len(labels)])
        for i, label in enumerate(labels):
            X_class = X[y.ravel()==label, :]
            Nc = X_class.shape[0]
            P[i] = Nc / N
            Mean[i, :] = X_class.mean(axis=0)
            Cov[:, :, i] = np.cov(X_class.T) + self.a*np.eye(d)
        self.Mean = Mean
        self.Cov = Cov
        self.P = P
    
    def predict(self, X):
        nC = len(self.P)
        L = np.zeros([nC, len(X)])
        for i in range(nC):
            L[i, :] = self.Q(X, self.Mean[i], self.Cov[:, :, i], self.P[i])
        yp = np.argmax(L, axis=0)
        return yp
    
    def accuracy(self, X, y):
        yp = self.predict(X)
        N = len(y)
        Nerror = (yp != y.ravel()).sum()
        print(1-Nerror/N)
        return 1-Nerror/N
```
